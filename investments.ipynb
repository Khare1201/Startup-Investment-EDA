```python
# Import pandas 
import pandas as pd 
import matplotlib.pyplot as plt

  
# reading csv file  
investments = pd.read_csv("investmentsVC.csv")
investments.head()
```

```python
investments.info()
```

```python
investments.describe()
```

```python
investments.isnull().sum()
```

```python
# we can sum() the figures in last 8 columns 
# so that we can get rid of few columns
# which are making out dataset too crowded

investments['round funding']= investments.iloc[:, -9:].sum(axis=1)
investments = investments.drop(columns = ['round_A', 'round_B', 'round_C', 'round_D', 'round_E', 'round_F', 'round_G', 'round_H'])
```

```python
# to get rid of date columns, as they are of very less importance for overall analysis
# funding_rounds, founded_at, founded_month, founded_quarter, founded_year
investments = investments.drop(columns =['funding_rounds', 'founded_at', 'founded_month', 'founded_quarter', 'founded_year'])
#investments.head()
```

```python
# getting few unique values to understand further
# uniquestat = investments['status'].unique()
#print(uniquestat)
#uniquemarket = investments['market'].unique()
#print(uniquemarket)
#uniquecc = investments['country_code'].unique()
#print(uniquecc)

# we can start by asking a question about what are the top market domains
# that can be approached for aquiring

market = investments.dropna(subset=['market']).reset_index()
c_by_market = market.groupby("market")["name"].count().sort_values(ascending=False)
#print(c_by_market)
X = c_by_market.head(8)
#print(c_by_market)
#market = market.groupby(['market', 'name'])
#market.head()
X.plot.bar()

plt.show()
```

```python
# It looks like our first plot was enough to analyze
# the basic market domain for recent startups
# Next step could be to examine basic funding trends
#investments.sum(axis = 0, skipna = True)

df = investments.drop(columns= ['permalink', 'homepage_url', 'category_list', 'state_code'])
df.head()
```

```python
c_by_city = df.groupby(["city", "market"])['name'].count().sort_values(ascending=False)
print(c_by_city)
```

```python
X = c_by_city.head(8)
X.plot.bar()
plt.show()
```

```python
# This tells that SF and NYC are the hotspot for latest tech innovations(obviously)
# we can look at the companies and their funding 
c_by_company = df.groupby(["name", "city"])['funding_total_usd'].max().sort_values(ascending=False)
c_by_company.head()
```
